```python
import pandas as pd
import os
```

```python
os.chdir("/ufrc/soltis/shan158538/TragFL_NewAnalysis/OutPut/DE_analysis/Additive_expression_Tms/")
```

```python
## Input files
nonAdditive = pd.read_table("DESeq2_Tms_nonAdditive_loci.txt", sep = "\t")
Additive = pd.read_table("DESeq2_Tms_Additive_loci.txt", sep = "\t")
nonDE = pd.read_table("DESeq2_noneDE_loci.txt", sep = "\t")
DE = pd.read_table("DESeq2_DE_loci.txt", sep = "\t")
```

```python
nonAdditive.head()
```

```python
## Intersections
nonAdditive_nonDE = pd.merge(nonAdditive, nonDE, how='inner', left_index=True, right_index=True)
nonAdditive_DE = pd.merge(nonAdditive, DE, how='inner', left_index=True, right_index=True)
Additive_nonDE = pd.merge(Additive, nonDE, how='inner', left_index=True, right_index=True)
Additive_DE = pd.merge(Additive, DE, how='inner', left_index=True, right_index=True)
```

```python
nonAdditive_nonDE.head()
```

```python
## Write out
## columns = [] means do not save any columns, but only save index
nonAdditive_nonDE.to_csv("Tms_nonAdditive_nonDE.txt", columns=[], header=False)
nonAdditive_DE.to_csv("Tms_nonAdditive_DE.txt", columns=[], header=False)
Additive_nonDE.to_csv("Tms_Additive_nonDE.txt", columns=[], header=False)
Additive_DE.to_csv("Tms_Additive_DE.txt", columns=[], header=False)
```
